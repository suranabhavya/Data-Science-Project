```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Columns of interest
categorical_cols = [
    'BDRM_COND', 'HEAT_SYSTEM', 'ORIENTATION', 'CORNER_UNIT', 'ROOF_STRUCTURE',
    'INT_WALL', 'EXT_FNISHED', 'INT_COND', 'EXT_COND', 'OVERALL_COND',
    'BTHRM_STYLE1', 'BTHRM_STYLE2', 'KITCHEN_TYPE', 'KITCHEN_STYLE1',
    'HEAT_TYPE', 'AC_TYPE', 'PROP_VIEW'
]

# Limit rare values and get top 5 categories for each column
top_n = 5
heatmap_data = pd.DataFrame()

for col in categorical_cols:
    top_categories = data_pa[col].value_counts(normalize=True, dropna=False).head(top_n)
    temp_df = pd.DataFrame(top_categories).T
    temp_df.index = [col]
    heatmap_data = pd.concat([heatmap_data, temp_df], axis=0)

# Fill NaNs with 0 (for categories not in top_n)
heatmap_data = heatmap_data.fillna(0)

# Plot
plt.figure(figsize=(16, 10))
sns.heatmap(heatmap_data, annot=True, fmt=".2f", cmap="Blues", cbar_kws={'label': 'Proportion'})
plt.title("Top Category Proportions Across Categorical Columns")
plt.ylabel("Column")
plt.xlabel("Category")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
cardinality = data_pa[categorical_cols].nunique(dropna=False).sort_values()

plt.figure(figsize=(10,6))
sns.barplot(x=cardinality.values, y=cardinality.index)
plt.title("Number of Unique Categories per Categorical Column")
plt.xlabel("Unique Values")
plt.ylabel("Column")
plt.tight_layout()
plt.show()
```

```python
from scipy.stats import entropy

entropy_scores = {}
for col in categorical_cols:
    counts = data_pa[col].value_counts(normalize=True, dropna=False)
    entropy_scores[col] = entropy(counts)

entropy_df = pd.Series(entropy_scores).sort_values()

plt.figure(figsize=(10,6))
sns.barplot(x=entropy_df.values, y=entropy_df.index)
plt.title("Entropy (Category Diversity) of Categorical Columns")
plt.xlabel("Entropy")
plt.ylabel("Column")
plt.tight_layout()
plt.show()
```
