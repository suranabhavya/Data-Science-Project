```python
import pandas as pd
import numpy as np
```

```python
pd.set_option('display.max_rows', None)  # Show all rows
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None) 
```

```python
violations_data = pd.read_csv('B:\BOSTON UNI\Acad\TDS\datasets\PWV_processed.csv')
sam_data = pd.read_csv('B:\BOSTON UNI\Acad\TDS\datasets\SAM.csv')
pad_data = pd.read_csv('B:\BOSTON UNI\Acad\TDS\datasets\PAD.csv')
```

```python
df_merged = violations_data.merge(sam_data[['SAM_ADDRESS_ID', 'PARCEL']], 
                        left_on='sam_id', 
                        right_on='SAM_ADDRESS_ID', 
                        how='left')

# Drop the redundant 'sam_Address_id' column
df_merged.drop(columns=['SAM_ADDRESS_ID'], inplace=True)
```

```python
df_merged.dropna(subset=['PARCEL'], inplace=True)
df_merged.isna().sum()
```

```python
violated_pids = df_merged['PARCEL'].unique()

pad_data['violation_bool'] = pad_data['PID'].isin(violated_pids).astype(int)
pad_data.head(n=10)
```

```python
# Summary of null values
null_summary = pad_data.isna().sum()
null_summary = null_summary[null_summary > 0].sort_values(ascending=False)
print("Null counts:\n", null_summary)

# Summary of zero or blank strings (for non-numeric)
zero_summary = (pad_data == 0).sum()
blank_summary = ((pad_data == "") | (pad_data == " ")).sum()

nonzero_info = pd.DataFrame({
    'Zero_Count': zero_summary,
    'Blank_Count': blank_summary
}).query("Zero_Count > 0 or Blank_Count > 0")

print("\nColumns with 0 or blanks:\n", nonzero_info)
```

```python
print("Dtypes and unique values:\n")
print(pad_data.dtypes)

print("\nExample row:")
print(pad_data.sample(1).T)
```

```python
cols_to_drop = [
    'RC_UNITS', 'RES_UNITS', 'COM_UNITS', #very sparsely populated
    'KITCHEN_STYLE1', 'KITCHEN_STYLE2', 'KITCHEN_STYLE3', # redundant info and also sparsely populated
    'BTHRM_STYLE1', 'BTHRM_STYLE2', 'BTHRM_STYLE3', #not useful for our analysis
    'STRUCTURE_CLASS', 'CORNER_UNIT', 'ORIENTATION', #more than 2/3 of the data is null and we think it is not useful
    'HEAT_SYSTEM', 'BDRM_COND', #redundant info and also sparsely populated
    'CD_FLOOR', 'UNIT_NUM','ST_NUM2', 'MAIL_ADDRESSEE' #junk columns
]
cleaned_data = pad_data.copy()
cleaned_data.drop(columns=cols_to_drop, inplace=True)

cleaned_data.head(n=10)
```

```python

cleaned_data.rename(columns={' GROSS_TAX ': 'GROSS_TAX'}, inplace=True)

#cleaning values like ' -\u2007\u2007 ' from the features
# chose these features as they are assinged a number to represent a sort of monetary value, but are encoded as objects.
val_cols = ['LAND_SF', 'LAND_VALUE', 'BLDG_VALUE', 'SFYI_VALUE', 'TOTAL_VALUE', 'GROSS_TAX']

for col in val_cols:

    cols = ( cleaned_data[col].astype(str).str.strip().replace(r'[^\d.\-]', '', regex=True).replace(r'^-$', '0', regex=True).replace(r'^$', '0', regex=True).astype(float) )

    try:
        cleaned_data[col] = cols.astype(float)
        print(f"Column {col} converted to float successfully.")
    except ValueError:

        print(f"Column {col} cannot be converted to float. Its Catergorical ")
```

```python
cleaned_data[['GROSS_AREA', 'LIVING_AREA', 'RES_FLOOR', 'FULL_BTH', 'KITCHENS']].describe()
```

```python
zero_to_nan = ['GROSS_AREA', 'LIVING_AREA', 'RES_FLOOR', 'FULL_BTH', 'KITCHENS']

for col in zero_to_nan:
    cleaned_data[col] = cleaned_data[col].replace(0, np.nan)

# we replace 0 with NaN because 0 doesn't make sense in the context of these features.
# we plan on using KNN imputation to fill in the missing values, so we need to convert 0 to NaN first.
```

```python
from sklearn.impute import KNNImputer

# Features for KNN: capture structure, size, layout
knn_features = [
    'GROSS_AREA', 'LIVING_AREA', 'RES_FLOOR', 'FULL_BTH',
    'HLF_BTH', 'KITCHENS', 'FIREPLACES', 'NUM_PARKING',
    'NUM_BLDGS', 'YR_BUILT', 'LAND_SF'
]


original_subset = cleaned_data[knn_features].copy()

imputer = KNNImputer(n_neighbors=5)
imputed_array = imputer.fit_transform(cleaned_data[knn_features])

# Update cleaned_data with imputed values
cleaned_data[knn_features] = imputed_array
```

```python
cleaned_data[knn_features].describe()
```

```python
# The max values for the features are not realistic, so we will cap them at a certain threshold.
cols_to_cap = [
    'GROSS_AREA', 'LIVING_AREA', 'RES_FLOOR',
    'FULL_BTH', 'KITCHENS', 'FIREPLACES',
    'NUM_PARKING', 'LAND_SF', 'YR_BUILT'
]

for col in cols_to_cap:
    upper = cleaned_data[col].quantile(0.99)
    lower = cleaned_data[col].quantile(0.01) if col in ['RES_FLOOR', 'YR_BUILT'] else 0  # only cap lower if needed
    cleaned_data[col] = cleaned_data[col].clip(lower=lower, upper=upper)

```

```python
# null value check
null_summary = cleaned_data.isna().sum()
null_summary = null_summary[null_summary > 0].sort_values(ascending=False)

if null_summary.empty:
    print("No missing values remain. Data is ready for encoding and modeling.")
else:
    print("Remaining nulls:\n", null_summary)
```

```python
# MORE COLS TO DROP

more_cols_to_drop = [
    'YR_REMODEL', 'CM_ID',
    'MAIL_STATE', 'MAIL_CITY', 'MAIL_ZIP_CODE', 'MAIL_STREET_ADDRESS', #not required
    'MAIL_ADDRESSEE', 'OWNER', 'ST_NUM' #owner might be useful, but its high caardinality makes it less useful for our analysis
]

cleaned_data.drop(columns=more_cols_to_drop, inplace=True, errors='ignore')
```

```python
#remaining categorical cols
cat_cols = cleaned_data.select_dtypes(include='object').columns
print("Categorical columns:\n", cat_cols)
cleaned_data[cat_cols] = cleaned_data[cat_cols].fillna('Unknown')
```

```python
num_cols = cleaned_data.select_dtypes(include=['float', 'int']).columns
print("Remaining numerical columns:\n", num_cols)
cleaned_data[num_cols] = cleaned_data[num_cols].fillna(cleaned_data[num_cols].median())
```
